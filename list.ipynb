```python
#  41: Merge and Sort Without sort()
list_a = [1, 3, 5, 7, 9]
list_b = [2, 4, 6, 8, 10]

merged = []
i = j = 0

# Merge lists while maintaining order
while i < len(list_a) and j < len(list_b):
    if list_a[i] < list_b[j]:
        merged.append(list_a[i])
        i += 1
    else:
        merged.append(list_b[j])
        j += 1

# Add remaining elements
while i < len(list_a):
    merged.append(list_a[i])
    i += 1
while j < len(list_b):
    merged.append(list_b[j])
    j += 1

print(f"Merged and sorted list: {merged}")
```

```python
# 42: Student Grade Calculator
students = [
    ["Alice", 85, 90, 78, 92],
    ["Bob", 88, 75, 82, 79],
    ["Charlie", 92, 95, 88, 94]
]

for student in students:
    name = student[0]
    grades = student[1:]
    average = sum(grades) / len(grades)
    
    if average >= 90:
        letter = "A"
    elif average >= 80:
        letter = "B"
    elif average >= 70:
        letter = "C"
    else:
        letter = "F"
    
    print(f"{name} - Average: {average:.2f}, Grade: {letter}")
```

```python
# 44: List Intersection Size

group_a = ["John", "Jane", "Bob", "Alice", "Charlie"]
group_b = ["Bob", "Charlie", "David", "Emma", "Frank"]

# Count common elements without creating a new list
count = 0
for person in group_a:
    if person in group_b:
        count += 1

print(f"Number of people in both groups: {count}")
```
