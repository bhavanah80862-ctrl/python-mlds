```python
#  41: Merge and Sort Without sort()
list_a = [1, 3, 5, 7, 9]
list_b = [2, 4, 6, 8, 10]

merged = []
i = j = 0

# Merge lists while maintaining order
while i < len(list_a) and j < len(list_b):
    if list_a[i] < list_b[j]:
        merged.append(list_a[i])
        i += 1
    else:
        merged.append(list_b[j])
        j += 1

# Add remaining elements
while i < len(list_a):
    merged.append(list_a[i])
    i += 1
while j < len(list_b):
    merged.append(list_b[j])
    j += 1

print(f"Merged and sorted list: {merged}")
```

```python
# 14: Find Common Elements in Two Lists
#list1 = [10, 20, 30, 40] and list2 = [30, 40, 50, 60], find the common elements using sets.

list1 = [10, 20, 30, 40]
list2 = [30, 40, 50, 60]

set_list1 = set(list1)
set_list2 = set(list2)

common_elements = set_list1.intersection(set_list2)
print(f"List 1: {list1}")
print(f"List 2: {list2}")
print(f"Common elements using sets: {common_elements}")
```

```python
# 15: Frozen Set
# Create a frozen set from a list.

my_list = [10, 20, 30]
frozen_set = frozenset(my_list)
print(f"Original list: {my_list}")
print(f"Created frozen set: {frozen_set}")
```

```python
# 16: Count Unique Words
#Write a code to count the number of unique words in the given a sentence.

sentence = "dog is a simple animal dogs is selfless animal"

words = sentence.lower().split()

unique_words = set(words)

unique_word_count = len(unique_words)

print(f"Original sentence: '{sentence}'")
print(f"Number of unique words: {unique_word_count}")
```

```python
#4.character counter

def count_characters(text):
    frequency = {}
    for char in text.replace(" ", ""):  # Ignore spaces
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency

# Example usage
print(count_characters("hello world"))
```
