```python
#(11) password strength checker

password = input("enter your password: ")
                 
has_upper = False
has_lower = False
has_digit = False
 
for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True

# Check all criteria
if len(password) >= 8 and has_upper and has_lower and has_digit:
    print("Strong password ")
else:
    print("Weak password ")

```

```python
#(12) Alternating case

text = input("Enter a string: ")

result = ""

for i, char in enumerate(text):
    if i % 2 == 0:
        result += char.upper()  
    else:
        result += char.lower() 

print(result)
```

```python
#(20) mask email/phone

contact = input("Enter an email or phone number: ")

if "@" in contact:
    username, domain = contact.split("@", 1)
    
    if len(username) > 2:
        masked_username = username[0] + "*" * (len(username)-2) + username[-1]
    else:
        masked_username = username[0] + "*"
    masked_contact = masked_username + "@" + domain
    print("Masked email:", masked_contact)


elif contact.isdigit() and len(contact) >= 4:
    masked_contact = "*" * (len(contact)-4) + contact[-4:]
    print("Masked phone number:", masked_contact)

else:
    print("Invalid input")
```

```python
# 11 : function returning tuple

def get_min_max(numbers):
    if not numbers:
        return (None, None)
    
    min_val = min(numbers)
    max_val = max(numbers)
    return (min_val, max_val)

my_numbers = [10, 5, 20, 2, 15]
min_max_values = get_min_max(my_numbers)
print(f"Original numbers: {my_numbers}")
print(f"Minimum and maximum values: {min_max_values}")
```

```python
#4.character counter

def count_characters(text):
    frequency = {}
    for char in text.replace(" ", ""):  # Ignore spaces
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    return frequency

# Example usage
print(count_characters("hello world"))
```
